## üìÅImport Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

## üìä Load Dataset & Data Cleaning

```python
df = pd.read_excel("man utd data.xlsx")
```

```python
df.dropna()
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Goals'] = pd.to_numeric(df['Goals'], errors='coerce')
df['Assist'] = pd.to_numeric(df['Assist'], errors='coerce')

df['xG'] = pd.to_numeric(df['xG'], errors='coerce')
df['xA'] = pd.to_numeric(df['xA'], errors='coerce')

df['G+A'] = df['Goals'] + df['Assist']
df['xG+xA'] = df['xG'] + df['xA']
```

## üìà Exploratory Data Analysis (EDA)

### Top 5 Players by Goals/Assists

```python
top_goals = df.nlargest(5, 'Goals')[['Player','Goals']]
sns.barplot(x='Goals', y='Player', data=top_goals,palette='Paired')
plt.title('Top 5 Players by Goals')
```

```python
top_assists = df.nlargest(5, 'Assist')[['Player','Assist']]
sns.barplot(x='Assist', y='Player', data=top_assists,palette='Paired')
plt.title('Top 5 Players by Assists')
```

## DISTRIBUTION PLOTS

```python
sns.histplot(df['Minutes'], bins=20, kde=True,color='#4CAF50')
plt.title('1. Distribution of Minutes Played (All Players)', fontsize=14)
plt.xlabel('Minutes')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
sns.countplot(x='Goals', data=df,palette='YlOrRd')
plt.title('2. Distribution of Total Goals Scored', fontsize=14)
plt.xlabel('Goals')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
df_filtered = df[df['Minutes'] > 450]
sns.kdeplot(df_filtered['Sh90'], fill=True,color='#00796B', alpha=0.6, linewidth=2)
plt.title('3. Density of Shots per 90 (Sh90) (Minutes > 450)', fontsize=14)
plt.xlabel('Shots per 90')
plt.grid(axis='y', linestyle='--')
plt.show()
```

## COMPARISON/RANKING PLOTS

```python
top_ga = df.sort_values('G+A', ascending=False).head(10)
sns.barplot(x='G+A', y='Player', data=top_ga,palette='viridis')
plt.title('Top 10 Players by Total Goals + Assists', fontsize=14)
plt.xlabel('Goals + Assists')
plt.ylabel('Player') 
plt.show()
```

```python
top_xga = df.sort_values('xG+xA', ascending=False).head(10)
sns.barplot(x='xG+xA', y='Player', data=top_xga, palette='magma')
plt.title('Top 10 Players by xG + xA (Expected Production)', fontsize=14)
plt.xlabel('Expected Goals + Expected Assists')
plt.ylabel('Player')
plt.show()
```

```python
sns.boxplot(x='Positon', y='Goals', data=df,palette='Set2')
plt.title('Goals Distribution by Position', fontsize=14)
plt.xlabel('Position')
plt.ylabel('Goals')
plt.show()
```

```python
sns.scatterplot(x='Appearances', y='Minutes', data=df,hue='Positon', palette='bright', s=150, edgecolor='k')
plt.title('Appearances vs. Minutes Played', fontsize=14)
plt.xlabel('Appearances')
plt.ylabel('Minutes')
plt.show()
```

## RELATIONSHIP PLOTS (Performance vs. Expectation)

```python
sns.scatterplot(x='xG', y='Goals', data=df,hue='Player', size='Minutes', sizes=(30, 400), legend=False, alpha=0.7)
plt.plot([df['xG'].min(), df['xG'].max()], [df['xG'].min(), df['xG'].max()], 'r--', label='Goal = xG')
plt.title('Goals vs. xG (Size = Minutes Played)', fontsize=14)
plt.xlabel('Expected Goals (xG)')
plt.ylabel('Actual Goals')
plt.show()
```

```python
sns.scatterplot(x='xA', y='Assist', data=df,hue='Player', size='Minutes', sizes=(30, 400), legend=False, alpha=0.7)
plt.plot([df['xA'].min(), df['xA'].max()], [df['xA'].min(), df['xA'].max()], 'r--', label='Assist = xA')
plt.title('Assists vs. xA (Size = Minutes Played)', fontsize=14)
plt.xlabel('Expected Assists (xA)')
plt.ylabel('Actual Assists')
plt.show()
```

```python
sns.regplot(x='Sh90', y='xG90', data=df_filtered,scatter_kws={'alpha':0.6, 'color':'#2C3E50'}, line_kws={'color':'#E74C3C', 'lw': 2})
plt.title('Sh90 vs. xG90 Correlation (Minutes > 450)', fontsize=14)
plt.xlabel('Shots per 90')
plt.ylabel('xG per 90')
plt.show()
```

```python
sns.regplot(x='KP90', y='xA90', data=df_filtered,scatter_kws={'alpha':0.6, 'color':'#3498DB'}, line_kws={'color':'#F39C12', 'lw': 2})
plt.title('KP90 vs xA90 Correlation (Minutes > 450)', fontsize=14)
plt.xlabel('Key Passes per 90')
plt.ylabel('xA per 90')
plt.show()
```

## Correlation Heatmap of Rate Stats

```python
correlation_matrix = df_filtered[['Goals', 'Assist', 'Sh90', 'KP90', 'xG90', 'xA90']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5,cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Heatmap of Key Rate Stats', fontsize=14)
plt.show()
```
